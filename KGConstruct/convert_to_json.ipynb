```python
import glob
import os
import re
import json

model = 'llama31-70b'
directory = f'/home/zhuxy/OSAutoFormalizer/KGConstruct/output-{model}'

# 使用glob遍历目录下的所有.txt文件
for filename in glob.glob(os.path.join(directory, '*.txt')):
    title = os.path.basename(filename)[:-4]
    with open (filename, 'r') as f:
        file_content = f.read()
    contents = file_content.split('Depth: 0, Query:')
    all_parent = []
    for content in contents[1:]:
        content = content.strip()
        content = content.split('\n')
        pattern = r"Current Concept to specify: (?P<concept>.+)$"
        match = re.search(pattern, content[0])
        query = match.group("concept")
        print(query)
        childrens = []
        for child in content[2:]:
            child_json = {}
            child_json['name']=child
            child_json['children']=[]
            childrens.append(child_json)    
        parent_json = {}
        parent_json['name']=query
        parent_json['children']=childrens
        all_parent.append(parent_json)

    if len(all_parent)==1:
        concept = all_parent[0]
    else:
        concept = {}
        concept['name'] = title
        concept['children'] = all_parent
    with open(f'/home/zhuxy/OSAutoFormalizer/KGConstruct/tree/{model}/{title}.txt', 'w') as f:
        json.dump(concept, f, indent=4)
        
```
